```python
# Importing Libraries
from pycaret.regression import *
import pandas as pd

# Download dataset from Kaggle and upload it to Colab or local environment
# Load Dataset
df = pd.read_csv('/content/train.csv')
```

```python
# Setup PyCaret for regression
reg = setup(data=df, target='SalePrice', session_id=123, use_gpu=True)
```

```python
# Compare models
best_model = compare_models()
```

```python
# Finalize model
final_model = finalize_model(best_model)
```

```python
# Predict on test data
predictions = predict_model(final_model, data=df)
```

```python

# Evaluate Model
evaluate_model(final_model)
```
